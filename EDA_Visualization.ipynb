```python
df[df["continent"].str.contains("Oceania")].sort_values(by="area_(km²)", ascending = False)
```

```python
#reversing the coloms so well get more linear data
cols = df.columns[5:13]
rev = cols[::-1]

dfv= df.groupby("continent")[rev].mean().sort_values(by="2020_population")

dfv
df3 = dfv.transpose()
df3
```

```python

# plt.figure(figsize=(20, 7))
# plt.plot(df3)
#easy way .plot is pandas function i mean plot is matplots plotting but pandas made it easy

df3.plot(figsize=(20,7))
```

```python
#data visaulizations

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("ice_cream_ratings.csv")
df = df.set_index("Date")
df
```

```python
# plt.style.use("dark_background")


df.plot(kind="line", subplots= True, title = "Ice Cream Ratings", xlabel = "Date", ylabel= "ratings")
```

```python
#barplot
df.plot(kind = "bar", stacked = True)
# df["Flavor Rating"].plot(kind = "bar", stacked = True)  it only shows the flavor rating bar graph
```

```python
#same thing with horizontal
df.plot.barh(stacked = True)
```

```python

plt.style.use("fivethirtyeight")
df.plot.scatter(x = "Flavor Rating", y = "Overall Rating", s = 100, c= "Black")
#this works with x and y defined. c = color and s = size are optional
```

```python
#histogram
df.plot.hist(bins = 20)
```

```python
#box plot
df.boxplot()
```

```python
#Area plot
df.plot.area(figsize=(10,5))
```

```python
plt.style.use('seaborn-v0_8')
df.plot.pie(y = "Flavor Rating", figsize= (10,10)) #required 1 argument
```
